```python
top_10_countries = jobs['company_location'].value_counts(normalize = True).head(10)
top_10_countries
```

```python
df = pd.DataFrame(index =jobs['work_year'].unique())

for year in df.index:
    total = 0
    for country in top_10_countries.index:
        df.loc[year,country] = jobs[(jobs['company_location'] == country) & (jobs['work_year'] == year)].shape[0]
        total += df.loc[year,country]

    df.loc[year,'Others'] = jobs[jobs['work_year'] == year].shape[0] - total

df
```

```python
fig, axes = plt.subplots(2,1, figsize = (10,10), sharex=True)

fig.suptitle ('salary growth analysis', fontsize = 20)

sns.set_style('darkgrid')
sns.lineplot(
    ax = axes[0],
    data = jobs,
    x = 'work_year',
    y = 'salary_in_usd',
    hue = 'experience_level',
    estimator='mean',
    errorbar= None
)

sns.lineplot(
    ax = axes[1],
    data = jobs,
    x = 'work_year',
    y = 'salary_in_usd',
    hue = 'job_category',
    estimator='mean',
    errorbar= None

)
plt.xticks(jobs['work_year'].unique())

plt.show()
```

```python
top_5_countries = jobs[jobs['work_year']==2022]['company_location'].value_counts(normalize = True).head(5)
top_5_countries
```

```python
sns.histplot(
    data = jobs[jobs['work_year']==2022],
    x = 'salary_in_usd',
    hue = 'experience_level',
    kde = True,
    bins = 20
)
```

```python
boxplot_dataframe = jobs[(jobs['work_year'] == 2022) & (jobs['company_location'].isin(top_5_countries.index))]
boxplot_dataframe
```

```python
medians = boxplot_dataframe.groupby('company_location')['salary_in_usd'].median().sort_values(ascending=False).index
medians
```

```python
plt.figure(figsize=(15,5))

sns.boxplot(
    data = boxplot_dataframe,
    x = 'company_location',
    y = 'salary_in_usd',
    hue = 'experience_level',
    order = medians
)
plt.show()
```
